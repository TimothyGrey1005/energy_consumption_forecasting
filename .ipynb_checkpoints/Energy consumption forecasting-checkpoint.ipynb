#### LSTM

```python
# Assuming 'daily_data' is already loaded with the necessary features and target
X = daily_data.drop(columns=['Consumption'])  # Features (exclude the target)
y = daily_data['Consumption']  # Target

# Train-Test Split (80/20 split, no shuffling for time series)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Scaling the features using MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Scale the target values separately
y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))

# Reshaping the input data for LSTM [samples, time steps, features]
X_train_lstm = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Import necessary libraries
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Function to evaluate model performance
def evaluate_model(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    return rmse, mse, mae

# Build the LSTM model before tuning
model_before_tuning = Sequential()
model_before_tuning.add(LSTM(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
model_before_tuning.add(LSTM(units=50, return_sequences=False))
model_before_tuning.add(Dense(25))
model_before_tuning.add(Dense(1))  # Output layer

# Compile the model
model_before_tuning.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model_before_tuning.fit(X_train_lstm, y_train_scaled, batch_size=32, epochs=20, verbose=1)

# Make predictions on the train and test data
y_train_pred_before = model_before_tuning.predict(X_train_lstm)
y_test_pred_before = model_before_tuning.predict(X_test_lstm)

# Rescale the predicted target values
y_train_pred_before_rescaled = y_scaler.inverse_transform(y_train_pred_before)
y_test_pred_before_rescaled = y_scaler.inverse_transform(y_test_pred_before)

# Evaluate on training data
rmse_train_before, mse_train_before, mae_train_before = evaluate_model(y_train, y_train_pred_before_rescaled)
print(f"Train RMSE before tuning: {rmse_train_before}")
print(f"Train MSE before tuning: {mse_train_before}")
print(f"Train MAE before tuning: {mae_train_before}")

# Evaluate on test data
rmse_test_before, mse_test_before, mae_test_before = evaluate_model(y_test, y_test_pred_before_rescaled)
print(f"Test RMSE before tuning: {rmse_test_before}")
print(f"Test MSE before tuning: {mse_test_before}")
print(f"Test MAE before tuning: {mae_test_before}")
```

#### LSTM + XGBoost

```python
# Assuming 'daily_data' is already loaded with the necessary features and target
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Step 1: Load and preprocess data
X = daily_data.drop(columns=['Consumption'])  # Features (exclude the target)
y = daily_data['Consumption']  # Target

# Train-Test Split (80/20 split, no shuffling for time series)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Scaling the features using MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Scale the target values separately
y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))

# Reshaping the input data for LSTM [samples, time steps, features]
X_train_lstm = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Import necessary libraries for LSTM
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Function to evaluate model performance
def evaluate_model(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    return rmse, mse, mae

# Step 2: Build the LSTM model to generate intermediate features
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
model_lstm.add(LSTM(units=50, return_sequences=False))
model_lstm.add(Dense(25))
model_lstm.add(Dense(1))  # Output layer

# Compile the model
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model_lstm.fit(X_train_lstm, y_train_scaled, batch_size=32, epochs=20, verbose=1)

# Generate LSTM predictions (intermediate features)
y_train_lstm_features = model_lstm.predict(X_train_lstm)
y_test_lstm_features = model_lstm.predict(X_test_lstm)

# Rescale the LSTM features back to the original scale
y_train_lstm_features_rescaled = y_scaler.inverse_transform(y_train_lstm_features)
y_test_lstm_features_rescaled = y_scaler.inverse_transform(y_test_lstm_features)

# Step 3: Train XGBoost using LSTM predictions as features
from xgboost import XGBRegressor

# Use LSTM predictions as features
X_train_xgb = y_train_lstm_features_rescaled
X_test_xgb = y_test_lstm_features_rescaled

# Train XGBoost model
model_xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6)
model_xgb.fit(X_train_xgb, y_train)

# Make predictions with the XGBoost model
y_train_pred_xgb = model_xgb.predict(X_train_xgb)
y_test_pred_xgb = model_xgb.predict(X_test_xgb)

# Step 4: Evaluate XGBoost model performance (after LSTM features)
rmse_train_xgb, mse_train_xgb, mae_train_xgb = evaluate_model(y_train, y_train_pred_xgb)
rmse_test_xgb, mse_test_xgb, mae_test_xgb = evaluate_model(y_test, y_test_pred_xgb)

print(f"Train RMSE (LSTM+XGBoost): {rmse_train_xgb}")
print(f"Train MSE (LSTM+XGBoost): {mse_train_xgb}")
print(f"Train MAE (LSTM+XGBoost): {mae_train_xgb}")

print(f"Test RMSE (LSTM+XGBoost): {rmse_test_xgb}")
print(f"Test MSE (LSTM+XGBoost): {mse_test_xgb}")
print(f"Test MAE (LSTM+XGBoost): {mae_test_xgb}")
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Step 1: Load and clean the dataset
daily_data = pd.read_csv('Complete_South_African_Energy_Consumption.csv')

# Convert the 'DateTime' column to datetime format
daily_data['DateTime'] = pd.to_datetime(daily_data['DateTime'], format='%m/%d/%y %I:%M %p', errors='coerce')

# Drop rows where 'DateTime' could not be converted or if there are NaNs in the numeric columns
daily_data = daily_data.dropna()

# Set 'DateTime' as the index (if you need to work with time series)
daily_data.set_index('DateTime', inplace=True)

# Step 2: Convert non-numeric entries to NaN in numeric columns
numeric_columns = ['Consumption', 'Production', 'Water', 'Wind', 'Hydroelectric', 'Oil and Gas', 'Coal', 'Solar']
for col in numeric_columns:
    daily_data[col] = pd.to_numeric(daily_data[col], errors='coerce')  # Convert invalid entries to NaN

# Drop rows with NaN values in the numeric columns
daily_data = daily_data.dropna()

# Step 3: Feature engineering and Preprocessing
X = daily_data.drop(columns=['Consumption'])  # Features (exclude the target)
y = daily_data['Consumption']  # Target

# Train-Test Split (80/20 split, no shuffling for time series)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Scaling the features using MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Scale the target values separately
y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))

# Reshape the input data for LSTM [samples, time steps, features]
X_train_lstm = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Step 4: Define and train LSTM model
def build_lstm_model(units=50, dropout_rate=0.2):
    model = Sequential()
    model.add(LSTM(units=units, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
    model.add(LSTM(units=units, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))  # Output layer
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Train the LSTM model
lstm_model = build_lstm_model(units=100)
lstm_model.fit(X_train_lstm, y_train_scaled, epochs=20, batch_size=64, verbose=1)

# Generate LSTM Features
y_train_lstm_features = lstm_model.predict(X_train_lstm)
y_test_lstm_features = lstm_model.predict(X_test_lstm)

# Step 5: Train Random Forest model using LSTM-generated features
best_rf_model = RandomForestRegressor(n_estimators=100, max_depth=5, min_samples_split=4, random_state=42)
best_rf_model.fit(y_train_lstm_features, y_train)

# Predict with the RandomForest model on LSTM features
y_train_pred_rf = best_rf_model.predict(y_train_lstm_features)
y_test_pred_rf = best_rf_model.predict(y_test_lstm_features)

# Step 6: Rescale predictions back to original scale
y_train_pred_rf_rescaled = y_scaler.inverse_transform(y_train_pred_rf.reshape(-1, 1))
y_test_pred_rf_rescaled = y_scaler.inverse_transform(y_test_pred_rf.reshape(-1, 1))

# Debugging: Check first few actual and predicted values to see if scaling is correct
print("Actual Consumption (first 10 values):", y_test.values[:10])
print("Predicted Consumption (first 10 values):", y_test_pred_rf_rescaled[:10])

# Step 7: Plot the Actual vs Predicted values
def plot_actual_vs_predicted(y_test, y_pred, model_name):
    plt.figure(figsize=(10, 6))
    plt.plot(np.arange(len(y_test)), y_test, label='Actual', color='blue', alpha=0.6)
    plt.plot(np.arange(len(y_pred)), y_pred, label=f'Predicted by {model_name}', color='red', alpha=0.6)
    plt.title(f'Actual vs Predicted Energy Consumption ({model_name})')
    plt.xlabel('Samples')
    plt.ylabel('Energy Consumption')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot actual vs predicted values
plot_actual_vs_predicted(y_test.values, y_test_pred_rf_rescaled, 'Random Forest + LSTM')

# Step 8: Residual analysis to see how far off the predictions are
residuals = y_test.values - y_test_pred_rf_rescaled.flatten()  # Flatten the predicted values
plt.figure(figsize=(10, 6))
plt.scatter(np.arange(len(residuals)), residuals, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residual Plot: Actual vs Predicted Energy Consumption')
plt.xlabel('Samples')
plt.ylabel('Residuals (Actual - Predicted)')
plt.grid(True)
plt.show()

# Step 9: Error distribution analysis (Optional)
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, bins=30)
plt.title('Error Distribution for Random Forest + LSTM')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
